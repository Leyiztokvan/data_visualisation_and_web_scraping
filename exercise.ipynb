```python
# Answering the question of Ex 2.5:
# You will notice that you only got 10 articles. Why aren't there more?
    # Because as stated in the API documentation under https://developer.nytimes.com/docs/articlesearch-product/1/overview, paragraph 'Pagination'
    # The Article Search API returns a max of 10 results at a time.

# Find out how many articles you should have found according to your API response.
    # The meta node in the response contains the total number of matches ("hits") and the current offset. 
    # Use the page query parameter to paginate through results (page=0 for results 1-10, page=1 for 11-20, ...). 
    # You can paginate thru up to 100 pages (1,000 results). If you get too many results try filtering by date range.
```

```python
# method to collect all articles (without time delay between requests)
def fetch_all_articles(base_url : str, search_params : dict) -> list :
    # List to store all articles
    all_articles = []
    # loop through the articles and collect all hits (same number of articles as number of hits)
    while True:
        # Make the request to the API
        resp = requests.get(base_url, params=search_params)
        
        # Check if the request was successful (status code 200)
        if resp.status_code == 200:
            # Extract articles from the response and add them to the list
            atcls = resp.json()['response']['docs']
            all_articles.extend(atcls)
            
            # Check if there are more pages to fetch
            hits = resp.json()['response']['meta']['hits']
            if len(all_articles) >= hits:
                break  # Stop pagination if we have collected all articles
            else:
                # Increment the page number for the next request
                search_params['page'] += 1
        else:
            # Print an error message if the request was not successful
            print('Error:', resp.status_code)
            break
    return all_articles
```

```python
# method to collect all articles (with time delay between requests)

# Note: Time module must be imported beforehand!
import time  # Import the time module

def fetch_all_articles_with_delay(base_url : str, search_params : dict, time_delay : int) -> list :
    # List to store all articles
    all_articles = []

    # loop through the articles and collect all hits (same number of articles as number of hits)
    while True:
        # Make the request to the API
        resp = requests.get(base_url, params=search_params)
        
        # Check if the request was successful (status code 200)
        if resp.status_code == 200:
            # Extract articles from the response and add them to the list
            atcls = resp.json()['response']['docs']
            all_articles.extend(atcls)
            
            # Check if there are more pages to fetch
            hits = resp.json()['response']['meta']['hits']
            if len(all_articles) >= hits:
                break  # Stop pagination if we have collected all articles
            else:
                # Increment the page number for the next request
                search_params['page'] += 1
                time.sleep(time_delay)  # Add a delay of 12 second between each request

        else:
            # Print an error message if the request was not successful
            print('Error:', resp.status_code)
            break
    return all_articles
```
